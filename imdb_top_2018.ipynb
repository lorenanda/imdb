```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
# Download IMDb's most popular movies in 2018
url = 'https://www.imdb.com/search/title?year=2018'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
type(soup)
```

```python
# Get the title
title = soup.title
print(title)
```

```python
movie_containers = soup.find_all('div', class_ = 'lister-item-content')
movie_containers
```

```python
len(movie_containers)
```

```python
movie_titles = soup.find_all('h3', class_ = 'lister-item-header')
movie_titles
```

```python
runtimes = soup.find_all('span', class_ = 'runtime')
runtimes
```

```python
genres = soup.find_all('span', class_ = 'genre')
genres
```

```python
values = soup.find_all('span', attrs={'name': 'nv'})
values
```

```python
imdb = []
df = pd.DataFrame(imdb, columns=["Rank", "Title", "Values"])
print(df)
```
